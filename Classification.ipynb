```python
#Sayılsal olmayan değerleri label encoding kullanarak sayısal değerlere dönüştürdük
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
#Sınıflandırmada etkisi olmayan sütunları sildik
data = data.drop(['title','countries','languages','type','release_date','genre','plot','description'],axis=1)
categorical = ['companies1','companies2','companies3','distributors1','distributors2','distributors3',
               'director','writer','producer','plot_keyword1','plot_keyword2','plot_keyword3','plot_keyword4','plot_keyword5',
               'musics','actor1','actor2','actor3']
data[categorical]=data[categorical].apply(lambda x:le.fit_transform(x))
```

```python
#Sezon kategorisini en son sutuna taşıdık
data['season_ct'] = data['season_c']
data = data.drop('season_c',axis=1)
data.head()
```

```python
#Tahmin edeceğimiz dizide bilinmemesi gerektiği için bitiş yılı, sezon ve bölüm sayısı değerlerini çıkardık
data = data.drop('end_year',axis=1)
data = data.drop('seasons',axis=1)
data = data.drop('episodes',axis=1)

#Henüz bitmemiş ve bitmiş dizileri ayrı setlere böldük
train_data = data[data["finished"] == 1] 
test_data = data[data["finished"] == 0]

#Finished sutunuyla işimiz kalmadığı için onu da çıkardık
train_data = train_data.drop('finished',axis=1)
test_data = test_data.drop('finished',axis=1)

print("Train Data Size:", len(train_data))
print("Test Data Size", len(test_data))
```

```python
train_data.shape
```

```python
#Bitmiş dizilerden modelimizi eğiteceğimiz test ve train setlerini böldük

X = train_data.iloc[:, 0:49].values
y = train_data.iloc[:, 49].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0,stratify = y)
```

```python
#Verilerimizi aynı ölçüte taşımak için scale ettik
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

#RandomForestClassifier modelimizi eğitim setini kullanarak eğittik
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
#Oluşan modelin başarısını gözlemlemek için bir classification_report oluşturduk.
from sklearn.metrics import classification_report
cr = classification_report(y_test,y_pred)
print(cr)
```

```python
#Modelin tek bir dizide çalışmasını sağlayacak fonksiyon
def predictSingle(data,test):
    cols = ['release_year']
    dt = data.copy()
    #Tahmin edilecek dizinin yapım yılından sonra yapılmış dizileri eğitim kümesinden çıkartma işlemi
    dt[cols] = dt[dt[cols] <= test.iloc[0][0]][cols]
    dt = dt.dropna()
    Xt = dt.iloc[:, 0:49].values
    yt = dt.iloc[:, 49].values
    Xt = sc.fit_transform(Xt)
    #Model eğitimi
    classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 0)
    classifier.fit(Xt, yt)
    
    T = test.iloc[:, 0:49].values
    result = classifier.predict(T)
    #Sonuç return etme kısmı
    print(result)
    if result[0] == 0:
        return "Başarısız"
    elif result[0] == 1:
        return "Kısmen Başarılı"
    elif result[0] == 2:
        return "Başarılı"
    elif result[0] == 3:
        return "Kısmen Olgun"
    elif result[0] == 4:
        return "Olgun"
```

```python
#Henüz bitmemiş bazı diziler için tahminler

result = predictSingle(train_data,test_data.loc[[755],:])
print(data_a.loc[[755],'title'])
print(f'{result}\n')

result = predictSingle(train_data,test_data.loc[[151],:])
print(data_a.loc[[151],'title'])
print(f'{result}\n')

result = predictSingle(train_data,test_data.loc[[30],:])
print(data_a.loc[[30],'title'])
print(f'{result}\n')

result = predictSingle(train_data,test_data.loc[[531],:])
print(data_a.loc[[531],'title'])
print(f'{result}\n')

result = predictSingle(train_data,test_data.loc[[50],:])
print(data_a.loc[[50],'title'])
print(f'{result}\n')



#for i in test_data.index.values:
#    result = predictSingle(train_data,test_data.loc[[i],:])
#    print(data_a.loc[[i],'title'])
#    print(result)
```
