```python
#Yayınlanma yılını yıl-gün-ay olarak farklı sütünlara ayırdık

releaseYear = pd.to_datetime(df["release_date"],format='%Y-%m-%d', errors = "coerce")
df['year'] = releaseYear.dt.year
df['month'] = releaseYear.dt.month
df['day'] = releaseYear.dt.day

for i in df.index:
    if pd.isna(df['year'][i]):
        df.loc[i,'year'] = df.loc[i,'release_year']
        df.loc[i,'month'] = 1.0
        df.loc[i,'day'] = 1.0
```

```python
df['year'] = df['year'].astype('int')
df['month'] = df['month'].astype('int')
df['day'] = df['day'].astype('int')
```

```python
#Genre altında bulunan farklı türleri unique olanları ayırarak sütunlar haline getirdik
listG =[]
df1['genre'].str.split(',').values.tolist()
for i in df1["genre"]:
    listG.append(i)
```

```python
#Sezon, Bölüm ve Süre değerlerini kategorize ettik

episodes_categorized = [0 if int(i)==0 
                else 0 if int(i)>0 and int(i) < 5
                else 1 if int(i) < 10 
                else 2 if int(i) < 20 
                else 3 if int(i) < 30 
                else 4 if int(i) < 40 
                else 5 if int(i) < 50 
                else 6 if int(i) < 60
                else 7 
                for i in df1["episodes"]]

df1["episode_c"] = episodes_categorized
seasons_categorized = [ 0 if int(i) == 1 # Başarısız
                else 1 if int(i) == 2 # Kısmen Başarılı
                else 2 if int(i) < 5   # Başarılı
                else 3 if int(i) < 10 # Kısmen-Olgunlaşmış
                else 4 # Olgunlaşmış
                for i in df1["seasons"]]

df1["season_c"] = seasons_categorized
runtime_categorized = [    0 if int(i)==0
                else 1 if int(i)>0 and int(i) < 1200
                else 2 if int(i) < 1800
                else 3 if int(i) < 3600
                else 4 if int(i) < 5400
                else 5 if int(i) < 7200
                else 6
                for i in df1["runtime"]]

df1["runtime_c"] = runtime_categorized
```

```python
#NaN olan endyear değerleri yerine 0 yazdık
df2=df1
df1['end_year'] = df1['end_year'].fillna(value=0.0)
df1['end_year'] = df1['end_year'].astype('int')

#Endyear değeri dolu olanları finished stununa 1 yazarak işaretledik
finished = [1 if i != 0 else 0 for i in df1["end_year"]] 
df1["finished"] = finished
df1['finished']

```

```python
#Senaryolar içindeki stopwordleri çıkardıktan sonra en çok geçen 5 kelimeyi bulup bunları stünlar haline getirdik

from collections import Counter
most=[]
listP =[]
stop_words =['all', 'just', 'being', 'over', 'both','two','one','three', 'through', 'yourselves', 'its', 'before', 'herself', 'had', 'should', 'to', 'only', 'under', 'ours', 'has', 'do', 'them', 'his', 'very', 'they', 'not', 'during', 'now', 'him', 'nor', 'did', 'this', 'she', 'each', 'further', 'where', 'few', 'because', 'doing', 'some', 'are', 'our', 'ourselves', 'out', 'what', 'for', 'while', 'does', 'above', 'between', 't', 'be', 'we', 'who', 'were', 'here', 'hers', 'by', 'on', 'about', 'of', 'against', 's', 'or', 'own', 'into', 'yourself', 'down', 'your', 'from', 'her', 'their', 'there', 'been', 'whom', 'too', 'themselves', 'was', 'until', 'more', 'himself', 'that', 'but', 'don', 'with', 'than', 'those', 'he', 'me', 'myself', 'these', 'up', 'will', 'below', 'can', 'theirs', 'my', 'and', 'then', 'is', 'am', 'it', 'an', 'as', 'itself', 'at', 'have', 'in', 'any', 'if', 'again', 'no', 'when', 'same', 'how', 'other', 'which', 'you', 'after', 'most', 'such', 'why', 'a', 'off', 'i', 'yours', 'so', 'the', 'having', 'once']
for i in df1["plot"]:
    i= i.replace('.','')
    i= i.replace(',','')
    i= i.replace(':','')
    i= i.replace(';','')
    i= i.replace('!','')
    i= i.replace('?','')  
    x = i.lower().split()
    fx = [word for word in x if word not in stop_words]
    Count = Counter(fx)
    most.append(Count.most_common(5))
plot_key_1 =[]
plot_key_2 =[]
plot_key_3 =[]
plot_key_4 =[]
plot_key_5 =[]
for i in most:
    plot_key_1.append(i[0][0]) 
    plot_key_2.append(i[1][0])
    try:
        plot_key_3.append(i[2][0]) 
    except:
        plot_key_3.append(i[0][0])
    try:
        plot_key_4.append(i[3][0]) 
    except:
        plot_key_4.append(i[0][0])
    try:
        plot_key_5.append(i[4][0]) 
    except:
        plot_key_5.append(i[0][0])
    
df1['plot_keyword1'] = plot_key_1
df1['plot_keyword2'] = plot_key_2
df1['plot_keyword3'] = plot_key_3
df1['plot_keyword4'] = plot_key_4
df1['plot_keyword5'] = plot_key_5
```
